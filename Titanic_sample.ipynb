```python
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
training = pd.read_csv('C:/Users/Shail/Desktop/Datasets/titanic_data.csv', sep=',')
training.head(10)
```

```python
training['Cabin'].fillna(0).value_counts()
```

```python
training = training.drop(['Name', 'Ticket', 'Cabin'], axis=1)
training.head(10)
```

```python
training.info()
```

```python
training['Embarked'].value_counts().plot(kind='bar')
```

```python
training['Embarked'] = training['Embarked'].fillna('S')
training = training.fillna(0)
training.info()
```

```python
embarked_dummies = pd.get_dummies(training['Embarked'])
embarked_dummies.head(10)
```

```python
training = training.join(embarked_dummies)
training.head(10)
```

```python
sex_dummies = pd.get_dummies(training['Sex'])
training = training.join(sex_dummies)
training = training.drop(['Embarked', 'Sex'], axis=1)
training.head(10)
```

```python
training = training.drop(['S', 'male'], axis=1)
training.head(10)
```

```python
training.describe()
```

```python
training['Survived'].mean()
```

```python
training_data = training.values[:,2:]
training_target = training.values[:,1]
training_target[:5]
```

```python
test = pd.read_csv('C:/Users/Shail/Desktop/Datasets/test.csv', sep=',')
test = test.drop(['Name', 'Ticket', 'Cabin'], axis=1)
test['Embarked'] = test['Embarked'].fillna(method='ffill')
test = test.fillna(0)
embarked_dummies = pd.get_dummies(test['Embarked'])
test = test.join(embarked_dummies)
sex_dummies = pd.get_dummies(test['Sex'])
test = test.join(sex_dummies)
test = test.drop(['Embarked', 'Sex'], axis=1)
test = test.drop(['S', 'male'], axis=1)
test_data = test.values
test_data[:3]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators = 1000)
rf_model.fit(training_data, training_target)
print(rf_model)
```

```python
expected = training_target
predicted = rf_model.predict(training_data)
print(metrics.accuracy_score(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print(metrics.roc_auc_score(expected, predicted))
```

```python
from sklearn import cross_validation
from sklearn.linear_model import LogisticRegression

num_folds = 10
num_instances = len(training_data)
kfold = cross_validation.KFold(n=num_instances, n_folds=num_folds)
model = LogisticRegression()
results = cross_validation.cross_val_score(model, training_data, training_target, cv=kfold)
print(results)
print(results.mean())
print(results.std())
```

```python
predicted = rf_model.predict(test_data[:,1:])
submission = np.c_[test_data[:,0].astype(int), predicted.astype(int)]
submission = pd.DataFrame(submission, columns=['PassengerId', 'Survived'])
submission.to_csv('C:/Users/Shail/Desktop/Datasets/submission_rf.csv', index=False)
submission.head()
```
